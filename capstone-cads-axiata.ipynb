### 2.2.1 Why columns such as `Cust_Last_Purchase` are `object` while they should be `float64`? Fix the type of the columns as it should be.



**Attention:** Some numerical columns have missing values, Dollar sign, or Comma. You need to fix the issue to be able to convert the column to numerical. 



**Hint:** 

1. The following code can help you to remove an 'OldSign' and replace it with a 'NewSign' or nothing: `df.col=df.col.str.replace('OldSign', 'NewSign')`



2. After removing the signs and replace it with correct sign, or nothing you need to:



    a- Create a list of the name of the categorical columns and the numerical columns:

        `CatCols=[Name of the Categorical columns]`

        `NumCols=list(set(data.columns)-set(CatCols))`

    

    b- Fix the type of the columns

        `data[CatCols] = data[CatCols].apply(lambda x: x.astype('category'))`

        `data[NumCols] = data[NumCols].apply(lambda x: x.astype('float64'))`

```python
# List of columns to clean and convert to numeric

columns_to_clean = [
    'Cust_Last_Purchase', 'Pur_latest', 'Pur_3_years_Avg', 
    'Pur_5_years_Avg', 'Pur_3_years_Avg_Indirect', 'Cust_Ann_Income'
]



# Remove dollar signs, commas, and other unwanted characters

for col in columns_to_clean:
    data[col] = data[col].str.replace('$', '').str.replace(',', '')



# Convert columns to numeric, setting errors='coerce' to handle any non-numeric values
data[columns_to_clean] = data[columns_to_clean].apply(pd.to_numeric, errors='coerce')



# Separate categorical and numerical columns
# Classify categorical columns based on the given column data

CatCols = [

    'Potential_Customer', 'C_ID', 'Status_Cust', 'Gender', 'Cust_Prop'

]

NumCols = list(set(data.columns) - set(CatCols))



# Convert categorical columns to 'category' type and numerical columns to 'float64'
data[CatCols] = data[CatCols].apply(lambda x: x.astype('category'))
data[NumCols] = data[NumCols].apply(lambda x: x.astype('float64'))



# Recheck the type of each column
print("Data types after conversion:")
print(data.dtypes)
```

1. How many categories in each categorical variables?

**Hint:**

For visualization you can use `sns.countplot()` for each categorical variable


```python
# 3.3 Explore categorical variables

for col in CatCols:
    # Count categories
    num_categories = data[col].nunique()
    print(f"\nColumn '{col}' has {num_categories} categories.")

    # Calculate proportions / percentage
    category_proportions = data[col].value_counts(normalize=True) * 100
    print(f"Proportion or Percentage of each category in '{col}':\n{category_proportions}")

    # Visualize with countplot
    plt.figure(figsize=(8, 4))
    ax = sns.countplot(data=data, x=col)
    plt.title(f"Distribution of Categories in '{col}'")
    plt.xlabel(col)
    plt.ylabel('Count')

    

    # Annotate each bar with the percentage
    for p in ax.patches:
        height = p.get_height()
        total = len(data)  # Total number of entries for calculating percentage
        proportion = f'{(height / total * 100):.1f}%'  # Calculate percentage
        ax.annotate(proportion, 

                    (p.get_x() + p.get_width() / 2., height),

                    ha='center', va='bottom')  # Center text and place above bar


    plt.show()
```

## 3.3 Find and Remove/Impute the Missing Values

```python
# Function to impute missing values

def impute_missing_values(df):

    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:  # Numerical columns
            if df[col].skew() > 1 or df[col].skew() < -1:  # Check for skewness
                # Use median for skewed distributions
                median_value = df[col].median()
                df[col].fillna(median_value, inplace=True)
                print(f"Filled missing values in '{col}' with median: {median_value}")

            else:
                # Use mean for non-skewed distributions
                mean_value = df[col].mean()
                df[col].fillna(mean_value, inplace=True)
                print(f"Filled missing values in '{col}' with mean: {mean_value}")

        else:  # Categorical columns
            mode_value = df[col].mode()[0]  # Fill with mode
            df[col].fillna(mode_value, inplace=True)
            print(f"Filled missing values in '{col}' with mode: {mode_value}")



# Impute missing values for training and test sets
impute_missing_values(X_train)
impute_missing_values(X_test)

# Check for missing values after imputation
print("")
print("Missing values after handling in X_train:")
print(X_train.isnull().sum())
print("")
print("Missing values after handling in X_test:")
print(X_test.isnull().sum())
```
